```python
# --- Step 4: Identify killer ---
# เงื่อนไข:
#   - เพศชาย
#   - member_id เริ่มต้นด้วย "48Z"
#   - membership_status = gold
#   - มีการ check-in วันที่ 20180115
killer_candidates = get_fit_now_member[
    (get_fit_now_member['id'].str.startswith("48Z")) &
    (get_fit_now_member['membership_status'] == "gold")
]

killer_checkins = get_fit_now_check_in[
    (get_fit_now_check_in['check_in_date'] == 20180115) &
    (get_fit_now_check_in['membership_id'].isin(killer_candidates['id']))
]

final_killer_suspect = killer_candidates.merge(
    killer_checkins, left_on="id", right_on="membership_id"
).merge(
    person, left_on="person_id", right_on="id"
)

print("=== ฆาตกร (final_killer_suspect) ===")
display(final_killer_suspect)

# --- Step 5: Identify mastermind ---
# เงื่อนไข:
#   - เพศหญิง
#   - สมาชิกยิม
#   - member_id = 48Z7A
mastermind_candidate = get_fit_now_member[get_fit_now_member['id'] == "48Z7A"]

mastermind_person_info = mastermind_candidate.merge(
    person, left_on="person_id", right_on="id"
)

print("=== ผู้บงการ (mastermind_person_info) ===")
display(mastermind_person_info)
```

```python
import pandas as pd

# --- โหลดข้อมูล ---
person = pd.read_csv('person.csv')
drivers_license = pd.read_csv('drivers_license.csv')
get_fit_now_member = pd.read_csv('get_fit_now_member.csv')

# --- ลบช่องว่างรอบๆ ชื่อคอลัมน์ ---
person.columns = person.columns.str.strip()
drivers_license.columns = drivers_license.columns.str.strip()
get_fit_now_member.columns = get_fit_now_member.columns.str.strip()

# --- ขั้นตอน 1: รวม drivers_license กับ person ---
suspects_with_info = pd.merge(
    drivers_license,
    person,
    left_on='id',        # id ใน drivers_license
    right_on='license_id',  # license_id ใน person
    suffixes=('_license', '_person')
)

# --- ขั้นตอน 2: รวมกับสมาชิกยิม ---
final_killer = pd.merge(
    suspects_with_info,
    get_fit_now_member,
    left_on='id_person',  # ตรวจสอบว่ามีคอลัมน์นี้จริงหรือเปล่า
    right_on='person_id',
    how='inner',
    suffixes=('_suspect', '_gym')
)

# --- ตรวจสอบทีละเงื่อนไข ---
print("=== Gender counts ===")
print(final_killer['gender'].value_counts())

print("\n=== Plate numbers containing 'H42W' ===")
print(final_killer[final_killer['plate_number'].str.contains('H42W', na=False)][['name_suspect','plate_number']])

print("\n=== Gold membership ===")
print(final_killer[final_killer['membership_status'].str.lower() == 'gold'][['name_suspect','membership_status']])

# --- กรองข้อมูลทั้งหมดพร้อมกัน ---
final_killer_filtered = final_killer[
    (final_killer['gender'].str.lower() == 'female') &
    (final_killer['plate_number'].str.contains('H42W', na=False)) &
    (final_killer['membership_status'].str.lower() == 'gold')
]

# --- แสดงผลลัพธ์สุดท้าย ---
print("\n=== ฆาตกรในคดีนี้ ===")
if final_killer_filtered.empty:
    print("ยังไม่มีใครตรงเงื่อนไขทั้งหมด")
else:
    print(final_killer_filtered[['name_suspect', 'plate_number', 'membership_status']])
```

```python
import pandas as pd

# --- โหลดข้อมูล ---
person = pd.read_csv('person.csv')
drivers_license = pd.read_csv('drivers_license.csv')
get_fit_now_member = pd.read_csv('get_fit_now_member.csv')

# --- ลบช่องว่างรอบๆ ชื่อคอลัมน์ ---
person.columns = person.columns.str.strip()
drivers_license.columns = drivers_license.columns.str.strip()
get_fit_now_member.columns = get_fit_now_member.columns.str.strip()

# --- รวม drivers_license กับ person ---
suspects_with_info = pd.merge(
    drivers_license,
    person,
    left_on='id',
    right_on='license_id',
    suffixes=('_license', '_person')
)

# --- รวมกับสมาชิกยิม ---
suspects_full = pd.merge(
    suspects_with_info,
    get_fit_now_member,
    left_on='id_person',
    right_on='person_id',
    how='left',  # ใช้ left join เพื่อดูคนที่อาจไม่มีสมาชิก
    suffixes=('_suspect', '_gym')
)

# --- เงื่อนไข 1: ผู้หญิง ---
female_suspects = suspects_full[suspects_full['gender'].str.lower() == 'female']
print("=== ผู้หญิง ===")
print(female_suspects[['name_suspect','gender']])

# --- เงื่อนไข 2: ป้ายทะเบียนมี 'H42W' ---
plate_suspects = suspects_full[suspects_full['plate_number'].str.contains('H42W', na=False)]
print("\n=== ป้ายทะเบียนมี 'H42W' ===")
print(plate_suspects[['name_suspect','plate_number']])

# --- เงื่อนไข 3: สมาชิก gold ---
gold_suspects = suspects_full[suspects_full['membership_status'].str.lower() == 'gold']
print("\n=== สมาชิก gold ===")
print(gold_suspects[['name_suspect','membership_status']])

# --- รวมทุกเงื่อนไขเพื่อหาฆาตกรตัวจริง ---
killer_final = suspects_full[
    (suspects_full['gender'].str.lower() == 'female') &
    (suspects_full['plate_number'].str.contains('H42W', na=False)) &
    (suspects_full['membership_status'].str.lower() == 'gold')
]

print("\n=== ฆาตกรตัวจริง ===")
if killer_final.empty:
    print("ยังไม่มีใครตรงเงื่อนไขทั้งหมด")
else:
    print(killer_final[['name_suspect','plate_number','membership_status']])
```
